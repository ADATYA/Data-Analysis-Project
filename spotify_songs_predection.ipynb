```python
most_5_tracks = df[["loudness","name"]].sort_values(by="loudness")
most_5_tracks
```

```python
most_5_tracks = df [["loudness" , "name"]].sort_values(by="loudness" , ascending=True)[:10]
most_5_tracks
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x="loudness",y="name" , data= most_5_tracks)
plt.title("Top Five loudest tracks")
plt.show()
```

```python
danceable = df[["danceability" , "name"]].sort_values(by="danceability")
danceable
```

```python
most_5_tracks= df[["danceability", "name"]].sort_values(by="danceability", ascending=True)[:5]
most_5_tracks
```

```python
plt.figure(figsize=(15, 9))
sns.barplot(x="danceability", y="name", data=most_5_tracks)
plt.title("Top five danceable tracks")
plt.show()
```

```python
top_10_instrumentalness_tracks= df[["instrumentalness", "name"]].sort_values(by="instrumentalness", ascending=False)[:10]
top_10_instrumentalness_tracks
```

```python
plt.figure(figsize=(15, 9))
plt.pie(x="instrumentalness", data=top_10_instrumentalness_tracks, autopct='%1.2f%%', labels=top_10_instrumentalness_tracks.name)
plt.show()
```
