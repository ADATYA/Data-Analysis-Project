### *Here isnull is not in same line,That's why I use here strip() function to remove leading whitespace*

```python
df.columns = df.columns.str.strip()
```

*sole unit are int value but this is show as a obj value couses ther is a doller signe, if the doller signe are removed then it is show the int value*


*So I should remove all doller signe and clean this data* .

```python
doller = ['Units Sold','Manufacturing Price','Sale Price','Gross Sales','Discounts','Sales','COGS','Profit']

for columns in doller:
  df[columns]=df[columns].str.replace('$', '' ,regex = True)
```

```python
df.head()
```

```python
df['Discounts']=df['Discounts'].replace('-','0', regex=True)

df.head()

# Other way to solce dass proble from Discounts :

#dass = ['Discounts']

#for columns in dass:
  #df[columns] = df[columns].str.replace('-','0', regex=True)
```

```python
# for converting the object datatype to floting valur
for columns in doller:
  df[columns] = df[columns].replace(',','', regex=True)

```

```python
df['Units Sold'] = df['Units Sold'].astype(float)
df['Manufacturing Price'] = df['Manufacturing Price'].astype(float)
df['Sale Price'] = df['Sale Price'].astype(float)
df['Gross Sales'] = df['Gross Sales'].astype(float)
df['Discounts'] = df['Discounts'].astype(float)
df['Sales'] = df['Sales'].astype(float)
df['COGS'] = df['COGS'].astype(float)
#['Units Sold','Manufacturing Price','Sale Price','Gross Sales','Discounts','Sales','COGS','Profit']
```

```python
(df['Profit']==' -  ').sum()
```

```python
df = df[df['Profit'] != ' -   ']  #
df['Profit'] = df['Profit'].str.strip()
df['Profit'] = df['Profit'].replace(r'[()]', '' , regex = True)
df['Profit'] = df['Profit'].astype(float)
df.dtypes
```

```python
# Coverting the date value using datetime keywords

df['Date']=pd.to_datetime(df['Date'])
df.dtypes
```

```python
df= df.drop(['Month Number','Month Name','Year',],axis=1)
df.dtypes
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

Now analysis the Country

```python
group = df.groupby('Country').agg({'Sales':'sum','Profit':'sum','Units Sold':'sum','COGS':'sum'}).reset_index()
```

```python
group.head(5)
```

```python
bar_width = 0.3
plt.figure(figsize=(10,5))
#Calculate the X position for the ber
x= np.arange(len(group['Country']))
plt.bar(x-bar_width, group['Sales'], bar_width, label='Sales') # label vul likhchilum :(

plt.xticks(x,group['Country'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
x= np.arange(len(group['Country']))
plt.bar(x-bar_width, group['Sales'], bar_width, label='Sales')
plt.bar(x,group['COGS'],bar_width,label='COGS')
# plt.bar(x+bar_width , group ['Units Sold'],bar_width, Label='Units Sold)

#plt.xlabel('Country')
#plt.xlabel('Values')
plt.xticks(x, group['Country'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.pie(group['Profit'], labels=group['Country'], autopct='%1.1f%%',startangle=140)
plt.title(' Country wise Profit Pie Chart ')
plt.show()
```

```python
group_product = df.groupby('Product').agg({'Units Sold':'sum','Manufacturing Price':'sum','Sale Price':'sum','Gross Sales':'sum','Profit':'sum','COGS':'sum'}).reset_index()
```
